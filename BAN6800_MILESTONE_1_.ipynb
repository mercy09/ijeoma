```python
import pandas as pd
import numpy as np
```

LOADING AND INTEGRATING DATA

```python
# Load datasets
orders = pd.read_csv('/content/olist_orders_dataset.csv')
customers = pd.read_csv('/content/olist_customers_dataset.csv')
order_items = pd.read_csv('/content/olist_order_items_dataset.csv')
products = pd.read_csv('/content/olist_products_dataset.csv')
payments = pd.read_csv('/content/olist_order_payments_dataset.csv')
reviews = pd.read_csv('/content/olist_order_reviews_dataset.csv')
```

```python
# Merge orders with customers
orders_customers = pd.merge(orders, customers, on='customer_id',  how='inner')

# Merge with order items
order_full = pd.merge(orders_customers, order_items, on='order_id', how='inner')

# Merge with product data
order_full = pd.merge(order_full, products, on='product_id', how='left')

# Merge with payment info
order_full = pd.merge(order_full, payments, on='order_id', how='left')
```

 STEP 2: CLEAN & PREPROCESSING DATA

```python
# Check missing and null values
missing = order_full.isnull().sum()
print(missing[missing > 0])
```

```python
order_full = order_full.dropna()
```

```python
order_full.isnull().sum()
```

```python
# Check total number of duplicate rows
duplicate_rows = order_full[order_full.duplicated()]
print("Number of duplicate rows:", duplicate_rows.shape[0])
```

```python
# Convert dates
date_cols = ['order_purchase_timestamp', 'order_approved_at',
             'order_delivered_carrier_date', 'order_delivered_customer_date',
             'order_estimated_delivery_date']

for col in date_cols:
    order_full[col] = pd.to_datetime(order_full[col], format='%d/%m/%Y %H:%M')
```

```python
order_full
```

STEP 3: TRANSFORMATION & FEATURE ENGINEERING

```python
#Delivery Time (actual vs estimated)
order_full['delivery_delay'] = (order_full['order_estimated_delivery_date'] - order_full['order_delivered_customer_date']).dt.days
order_full['actual_delivery_time'] = (order_full['order_delivered_customer_date'] - order_full['order_purchase_timestamp']).dt.days
```

```python
#Price-related Features
# Total order price per product
order_full['order_item_total'] = order_full['price'] + order_full['freight_value']
```

```python
#Order Count Per Customer
order_count = order_full.groupby('customer_id')['order_id'].nunique().reset_index()
order_count.columns = ['customer_id', 'total_orders']
order_full = pd.merge(order_full, order_count, on='customer_id', how='left')
```

```python
#Monetary Value Per Customer (Total Spend)
customer_spend = order_full.groupby('customer_id')['order_item_total'].sum().reset_index()
customer_spend.columns = ['customer_id', 'total_spent']
order_full = pd.merge(order_full, customer_spend, on='customer_id', how='left')
```

```python
#Recency Feature
import datetime
reference_date = order_full['order_purchase_timestamp'].max()
recency_df = order_full.groupby('customer_id')['order_purchase_timestamp'].max().reset_index()
recency_df['recency_days'] = (reference_date - recency_df['order_purchase_timestamp']).dt.days
order_full = pd.merge(order_full, recency_df[['customer_id', 'recency_days']], on='customer_id', how='left')
```
