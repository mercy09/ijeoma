RECREATING NEW AND IMPORTANT COLUMN

```python
dataset = dataset[['coaster_name','Location', 'Status','Manufacturer','year_introduced','latitude', 'longitude','Type_Main','opening_date_clean',
'speed_mph', 'height_ft', 'Inversions_clean', 'Gforce_clean;;;']].copy()
```

```python
dataset
```

CHECKING FOR NULL OR MISSING VALUES

```python
dataset.isnull().sum()
```

REMOVE NULL VALUES

```python
dataset = dataset.dropna()
```

```python
dataset
```

```python
dataset.isna().sum()
```

CHECK DUPLICATES

```python
dataset.loc[dataset.duplicated(subset=['Manufacturer'])].head(5)
```

DROP DUPLICATES

```python
dataset = dataset.loc[~dataset.duplicated(subset=['coaster_name','Location','opening_date_clean'])] \
    .reset_index(drop=True).copy()
```
