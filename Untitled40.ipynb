```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate manufacturing dataset
num_records = 200

data = {
    "Product_ID": np.arange(1, num_records + 1),
    "Cycle_Time_sec": np.random.normal(loc=120, scale=10, size=num_records).round(2),  # Normal distribution
    "Defect_Rate": np.random.uniform(0, 5, num_records).round(2),  # Percentage of defects (0-5%)
    "Temperature_C": np.random.normal(loc=200, scale=15, size=num_records).round(2),  # Normal distribution
    "Pressure_Bar": np.random.normal(loc=50, scale=5, size=num_records).round(2),  # Normal distribution
    "Operator_Skill_Level": np.random.randint(1, 6, num_records),  # Skill level from 1 to 5
    "Machine_Age_Years": np.random.randint(1, 11, num_records),  # Machine age from 1 to 10 years
    "Maintenance_Frequency": np.random.choice(["Low", "Medium", "High"], num_records),  # Maintenance category
    "Downtime_Minutes": np.random.normal(loc=30, scale=8, size=num_records).round(2),  # Normal distribution
    "Yield_Percentage": np.random.uniform(85, 100, num_records).round(2),  # Yield rate between 85-100%
}

# Create DataFrame
df = pd.DataFrame(data)

# Display first few rows
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Manufacturing Data")
plt.show()

# Scatter Plot: Cycle Time vs. Yield Percentage
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Cycle_Time_sec"], y=df["Yield_Percentage"], alpha=0.7)
plt.title("Cycle Time vs. Yield Percentage")
plt.xlabel("Cycle Time (sec)")
plt.ylabel("Yield Percentage")
plt.show()

# Scatter Plot: Defect Rate vs. Yield Percentage
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Defect_Rate"], y=df["Yield_Percentage"], alpha=0.7, color="red")
plt.title("Defect Rate vs. Yield Percentage")
plt.xlabel("Defect Rate (%)")
plt.ylabel("Yield Percentage")
plt.show()

# Boxplot: Maintenance Frequency vs. Yield Percentage
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Maintenance_Frequency"], y=df["Yield_Percentage"], palette="Set2")
plt.title("Maintenance Frequency vs. Yield Percentage")
plt.xlabel("Maintenance Frequency")
plt.ylabel("Yield Percentage")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Prepare features and target variable
X = df.drop(columns=["Product_ID", "Yield_Percentage"])  # Features
y = df["Yield_Percentage"]  # Target

# One-hot encode categorical variable
X = pd.get_dummies(X, columns=["Maintenance_Frequency"], drop_first=True)

# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_pred = rf_model.predict(X_test)

# Model Evaluation
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

rmse, r2
```
