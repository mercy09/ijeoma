```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the uploaded Boston dataset
df = pd.read_csv("/content/boston (1).csv")
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

SIMPLE LINEAR REGRESSION RESULTS

```python
# Set visual style
sns.set(style="whitegrid")
```

```python
# Store results
simple_regression_results = {}
```

```python
# Response variable
y = df["CRIM"]
y
```

```python
# Loop through each predictor to fit a simple linear regression
for col in df.columns:
    if col != "CRIM":
        X = sm.add_constant(df[col])  # add intercept
        model = sm.OLS(y, X).fit()
        simple_regression_results[col] = {
            "model": model,
            "p_value": model.pvalues[1],
            "coefficient": model.params[1],
            "r_squared": model.rsquared
        }
```

```python
# Extract results to a DataFrame for easier visualization
summary_df = pd.DataFrame.from_dict(simple_regression_results, orient='index')
summary_df_sorted = summary_df.sort_values(by="p_value")

summary_df_sorted[["coefficient", "p_value", "r_squared"]]
```

MULTIPLE LINEAR REGRESSION RESULTS

```python
# Prepare predictors and response for multiple regression
X_all = df.drop(columns="CRIM")
X_all = sm.add_constant(X_all)  # Add intercept
y = df["CRIM"]
```

```python
# Fit multiple linear regression
multi_model = sm.OLS(y, X_all).fit()
```

```python
# Extract p-values and coefficients
multi_results = pd.DataFrame({
    "coefficient": multi_model.params,
    "p_value": multi_model.pvalues
})
```

```python
# Drop the intercept for analysis
multi_results = multi_results.drop(index="const")
```

```python
# Sort by p-value for easier interpretation
multi_results_sorted = multi_results.sort_values(by="p_value")
multi_results_sorted
```

PLOTING UNIVARIATE VS MULTIVARIATE COEFFICIENTS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
# Prepare response variable
y = df["CRIM"]
```

```python
 #--- Step 1: Univariate regression coefficients ---
uni_results = {}
for predictor in df.columns.drop("CRIM"):
    X = sm.add_constant(df[predictor])
    model = sm.OLS(y, X).fit()
    uni_results[predictor] = model.params[predictor]
```

```python
# --- Step 2: Multivariate regression coefficients ---
X_multi = sm.add_constant(df.drop(columns="CRIM"))
multi_model = sm.OLS(y, X_multi).fit()
multi_results = multi_model.params.drop("const")
```

```python
# --- Step 3: Combine results ---
compare_df = pd.DataFrame({
    "univariate_coef": pd.Series(uni_results),
    "multivariate_coef": multi_results
})
```

```python
# --- Step 4: Plot ---
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="univariate_coef",
    y="multivariate_coef",
    data=compare_df,
    s=100,
    color="skyblue"
)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.axvline(0, color='gray', linestyle='--', linewidth=1)

# Add labels to points
for i in compare_df.index:
    plt.text(compare_df.loc[i, "univariate_coef"],
             compare_df.loc[i, "multivariate_coef"],
             i, fontsize=9, ha='right')

plt.title("Comparison of Univariate vs Multivariate Coefficients", fontsize=14)
plt.xlabel("Univariate Regression Coefficient", fontsize=12)
plt.ylabel("Multivariate Regression Coefficient", fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

FITTING A CUBIC POLYNOMIAL REGRESSION

```python
# Create a dataframe to store results
nonlinear_results = []
```

```python
# Response variable
y = df["CRIM"]
```

```python
# Loop over each predictor
for predictor in df.columns.drop("CRIM"):
    X = df[[predictor]]
    X_poly = pd.DataFrame({
        f"{predictor}": X[predictor],
        f"{predictor}^2": X[predictor] ** 2,
        f"{predictor}^3": X[predictor] ** 3
    })
    X_poly = sm.add_constant(X_poly)
```

```python
# Fit model
model = sm.OLS(y, X_poly).fit()
```

```python
# Store adjusted R² and p-values
nonlinear_results.append({
        "Predictor": predictor,
        "Adj R²": round(model.rsquared_adj, 4),
        "p-value (X²)": round(model.pvalues.get(f"{predictor}^2", float("nan")), 4),
        "p-value (X³)": round(model.pvalues.get(f"{predictor}^3", float("nan")), 4),
        "Non-linear?": (
            "Yes" if model.pvalues.get(f"{predictor}^2", 1) < 0.05 or
                     model.pvalues.get(f"{predictor}^3", 1) < 0.05 else "No"
        )
    })
```

```python
# Create DataFrame for easy viewing
nonlinear_df = pd.DataFrame(nonlinear_results).sort_values(by="Adj R²", ascending=False)
print(nonlinear_df.to_string(index=False))
```
