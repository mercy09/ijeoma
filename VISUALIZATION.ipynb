```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Sample Data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
```

```python
x

```

```python
y
```

```python
# Creating a Simple Line Plot
plt.figure(figsize= (8, 5))
plt.plot(x, y , label= 'Linear  Growth', color = 'red', linestyle = '-', linewidth = 2, marker ='o')
plt.title('Linear Growth')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#ADDING TITLE AND LABELS
plt.title('Simple line plot of Growth', fontsize = 16)
plt.xlabel('X-axis (Time)', fontsize = 14)
plt.ylabel('Y_axis (Value)', fontsize = 14)
plt.grid(True)
plt.plot(x, y, label = 'Linear Growth', color = 'blue', linestyle = '-', linewidth = 2, marker = 'o')
plt.legend()
plt.show()
```

SEABORN VISUALIZATION

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#LOAD DATASET
tips = sns.load_dataset('tips')
tips
```

```python
#CREATING A SCATTER PLOT
plt.figure(figsize = (10, 6))
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', style = 'time', s = 200)
```

```python
#ENHANCING THE PLOT
plt.figure(figsize = (10, 6))
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'day', style = 'time', s = 300)
plt.title('scatter plot of tips vs total bill', fontsize = 18)

plt.grid(True)
plt.show()
```

```python
sns.pairplot(tips, hue = 'day', diag_kind = 'kde')
```

```python
sns.histplot(data = tips, x = 'total_bill', hue = 'day', kde = True)
```
